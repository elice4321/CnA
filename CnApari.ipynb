```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input/mumbai-locals'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
df = pd.read_csv('/kaggle/input/mumbai-locals/Mumbai_Train_final.csv')
# Group by 'Platforms' and calculate the mean footfall
footfall_by_platforms = df.groupby('Platforms')['Footfall (lakhs)'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='Platforms', y='Footfall (lakhs)', data=footfall_by_platforms, palette='viridis')
plt.title('Average Footfall vs Number of Platforms')
plt.xlabel('Number of Platforms')
plt.ylabel('Average Footfall (in Lakhs)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Enable grid lines for the y-axis
plt.show()
# Group by 'Tracks' and calculate the mean footfall
footfall_by_tracks = df.groupby('Tracks')['Footfall (lakhs)'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='Tracks', y='Footfall (lakhs)', data=footfall_by_tracks, palette='viridis')
plt.title('Average Footfall vs Number of Tracks')
plt.xlabel('Number of Tracks')
plt.ylabel('Average Footfall (in Lakhs)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Enable grid lines for the y-axis
plt.ylim(0, 8)
plt.show()
# Convert distance to numeric (removing ' KM' and converting to float)
df['Distance From Previous of the Line'] = df['Distance From Previous of the Line'].str.replace(' KM', '').astype(float)
plt.figure(figsize=(12, 6))
sns.lineplot(x='Distance From Previous of the Line', y='Footfall (lakhs)', data=df, marker='o')
plt.title('Overall Footfall vs Distance from Previous Station (All Regions)')
plt.xlabel('Distance from Previous Station (KM)')
plt.ylabel('Footfall (in Lakhs)')
plt.grid(True)
plt.show()
# Convert time to numeric (removing ' min' and converting to integer)
df['Time taken From Previous of the Line'] = df['Time taken From Previous of the Line'].str.replace(' min', '').astype(int)
plt.figure(figsize=(12, 6))
sns.lineplot(x='Time taken From Previous of the Line', y='Footfall (lakhs)', data=df, marker='o')
plt.title('Footfall vs Time Taken from Previous Station (All Regions)')
plt.xlabel('Time Taken from Previous Station (Minutes)')
plt.ylabel('Footfall (in Lakhs)')
plt.grid(True)
plt.show()
# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```
