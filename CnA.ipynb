```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python



# Group by 'Line' and calculate the mean attraction popularity score
line_popularity = df.groupby('Line')['Attraction Popularity Score'].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(line_popularity['Line'], line_popularity['Attraction Popularity Score'], color='skyblue')
plt.title('Attraction Popularity by Line')
plt.xlabel('Line')
plt.ylabel('Attraction Popularity Score')
plt.xticks(rotation=45)
plt.show()
```

```python

# Assuming your DataFrame is named 'df'
# Count the number of stations per line
line_distribution = df['Line'].value_counts()

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(line_distribution, labels=line_distribution.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribution of Stations by Line')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python

sns.set(style="whitegrid")

# Plotting
plt.figure(figsize=(12, 6))
sns.boxplot(x='Line', y='Footfall (lakhs)', data=df, palette='Set2')
plt.title('Box Plot of Footfall by Line')
plt.xlabel('Line')
plt.ylabel('Footfall (lakhs)')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame creation (replace this with your actual DataFrame)
# df = pd.read_csv('your_data_file.csv')  # Uncomment this line to load your data

# Check the column names
print("Column names in the DataFrame:", df.columns)

# Clean column names by stripping spaces and converting to lowercase
df.columns = df.columns.str.strip()  # Remove leading/trailing spaces
df.columns = df.columns.str.lower()   # Convert to lowercase

# Grouping the data by station and summing the footfall
station_footfall = df.groupby('station')['footfall (lakhs)'].sum().reset_index()

# Plotting
plt.figure(figsize=(15, 8))
plt.bar(station_footfall['station'], station_footfall['footfall (lakhs)'], color='lightblue')
plt.title('Total Footfall by Station')
plt.xlabel('Station')
plt.ylabel('Total Footfall (lakhs)')
plt.xticks(rotation=90)  # Rotate station names for better visibility
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python


# Prepare the data for the scatter plot
x = df['Footfall (lakhs)']  # Footfall data
y = df['Attraction Popularity Score']  # Attraction popularity score

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.6)

# Adding titles and labels
plt.title('Footfall vs. Attraction Popularity Score', fontsize=16)
plt.xlabel('Footfall (in lakhs)', fontsize=14)
plt.ylabel('Attraction Popularity Score', fontsize=14)

# Adding grid for better readability
plt.grid(True)

# Show the plot
plt.show()
```

```python

# Group the data by Line and sum the footfall
footfall_by_line = df.groupby('Line')['Footfall (lakhs)'].sum().reset_index()

# Create the line graph
plt.figure(figsize=(10, 6))
plt.plot(footfall_by_line['Line'], footfall_by_line['Footfall (lakhs)'], marker='o', color='blue', linestyle='-', alpha=0.7)

# Adding titles and labels
plt.title('Total Footfall by Train Line', fontsize=16)
plt.xlabel('Train Line', fontsize=14)
plt.ylabel('Total Footfall (in lakhs)', fontsize=14)

# Adding grid for better readability
plt.grid(True)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```

```python

x = df['Footfall (lakhs)']
y = df['Attraction Popularity Score']

# Plot
plt.figure(figsize=(10, 6))
sns.regplot(x=x, y=y, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Footfall vs. Attraction Popularity Score')
plt.xlabel('Footfall (in lakhs)')
plt.ylabel('Attraction Popularity Score')
plt.grid(True)
plt.show()
```
